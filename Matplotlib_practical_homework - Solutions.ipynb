### P6. Write a Python code to output the following graph. (or a similar one). 
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

````python
g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
labels = ['M', 'T', 'W', 'Th', 'F']
index = np.arange(len(labels))
width = 0.2
plt.bar(index - width, g1, width, color='c', label = 'Group1')
plt.bar(index, g2, width, color='r', label = 'Group2')
plt.bar(index + width, g3, width, color='b', label = 'Group3')
plt.xlabel('Groups')
plt.ylabel('Productivity')
plt.title('Productivity during the week')
plt.xticks(index, labels)
plt.legend()
plt.show()
````

### P8. Plotly

```
Load the Iris dataset from plotly.express. 
```

````python
import plotly.express as px
ds = px.data.iris()
ds.head()
````

```
Create a scatter plot of petal_length vs petal_width. Use species as the color and display all the feature values when you hover over a point.
```

````python
fig = px.scatter(ds, x = 'petal_length', y = 'petal_width', color = ds['species'], hover_data = ds.columns)
fig.show()
````

```
Create a histogram having 20 bins using the feature sepal_length. Use species feature as the color. 
```

````python
fig1 = px.histogram(ds, x = 'sepal_length', nbins = 20, color = 'species')
fig2 = px.histogram(ds, x = 'sepal_length', nbins = 20, color = 'species', facet_row = ds['species'])
fig1.show()
fig2.show()
````

```
Plot sepal_length and sepal_width features on the same histogram. 
```

````python
fig = px.histogram(ds, ['sepal_length','sepal_width'], nbins = 40)
fig.show()
````

```
Create a heatmap showing the correlation of the columns sepal_width and petal_width.
```

````python
px.imshow(ds[["sepal_width", "petal_width"]].corr())
````

```
Create a box plot showing the distribution of the sepal_length feature. 
```

````python
px.box(ds.sepal_length)
````
